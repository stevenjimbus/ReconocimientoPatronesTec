# Paradigma Machine Learning

Etapas básicas de un sistema de ML

```python
display.Image("paradigmaMLtradicional.png")
```

**1. Pre-procesamiento**

Consiste en identificar partes o componentes del conjunto de datos que sean incompletas, imprecisas, incorrectas o irrelevantes, de manera tal que puedan ser reemplazadas, modicadas o removidas.

**2. Extraccion de caracteristicas**

La extracción de características comienza a partir de un conjunto inicial de datos medidos y construye valores derivados (características) destinados a ser informativos y no redundantes, facilitando los pasos posteriores de aprendizaje y generalización y, en algunos casos, conduciendo a mejores interpretaciones humanas.


La extracción de características es un término general para los métodos de construcción de combinaciones de las variables para solucionar estos problemas sin dejar de describir los datos con suficiente precisión. Muchos profesionales del aprendizaje automático creen que la extracción de características correctamente optimizada es la clave para una construcción de modelos eficaz. 
Recordar que para un sistema: If Trash In -> Trash out


**3. Investigación/Selección de Modelo - Clasificación/Regresión**

1. Un modelo predictivo trata sobre el problema de aprender una función de mapeo de entradas a salidas llamada aproximación de función.
2. Clasificación es el problema de predecir una salida de etiqueta de clase discreta para un ejemplo.
3. Regresión es el problema de predecir una producción de cantidad continua, por ejemplo.


**4. Validación**

En el aprendizaje automático, la validación del modelo se conoce como el proceso en el que un modelo entrenado se evalúa con un conjunto de datos de prueba. El conjunto de datos de prueba es una parte separada del mismo conjunto de datos del que se deriva el conjunto de entrenamiento. El objetivo principal de utilizar el conjunto de datos de prueba es probar la capacidad de generalización de un modelo entrenado


# Paradigma DevOPS

```python
display.Image("paradigmaDevOPS.png",width=600)
```

1. **Plan:** esta es la parte del proyecto en la que está organizando las tareas, los horarios y configura las herramientas de gestión de proyectos.
2. **Code:** Aquí los desarrolladores de software están desarrollando y revisando el código.
3. **Build:** este paso es el primero hacia la automatización. El objetivo aquí es construir el código fuente en un formato deseado, compilando, probando y desplegando en un lugar particular de la infraestructura.
4. **Test:** El proceso de prueba continuo reduce los riesgos. Las pruebas automáticas aseguran que no se implementarán errores en producción.
5. **Release:** El código ha pasado el proceso de prueba (integración continua) y está listo para implementarse.
6. **Deploy:** El equipo operativo está implementando la nueva función en producción. Pero como la automatización es uno de los principios de DevOps, es posible configurar una implementación continua.
7. **Operate:** Las operaciones construyen o mantienen una infraestructura escalable
8. **Monitor:** El monitoreo es un paso importante ya que permite solucionar incidentes más rápido y crear una mejor experiencia para su usuario final.

# Paradigma MLOPS

```python
display.Image("MLOps-Neal-Analytics.png",width=600)
```

MLOps se basa en la disciplina existente de DevOps, la práctica moderna de escribir, implementar y ejecutar aplicaciones empresariales de manera eficiente. DevOps comenzó hace una década como una forma en que tribus en guerra de desarrolladores de software (Devs) y equipos de operaciones de TI (Ops) podían colaborar.

MLOps agrega al equipo a los científicos de datos, quienes seleccionan conjuntos de datos y construyen modelos de IA que los analizan. También incluye ingenieros de ML, que ejecutan esos conjuntos de datos a través de los modelos de forma disciplinada y automatizada.

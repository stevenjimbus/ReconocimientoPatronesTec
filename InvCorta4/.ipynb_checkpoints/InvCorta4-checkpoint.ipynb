**2. Extraccion de caracteristicas**

La extracci√≥n de caracter√≠sticas comienza a partir de un conjunto inicial de datos medidos y construye valores derivados (caracter√≠sticas) destinados a ser informativos y no redundantes, facilitando los pasos posteriores de aprendizaje y generalizaci√≥n y, en algunos casos, conduciendo a mejores interpretaciones humanas.


La extracci√≥n de caracter√≠sticas es un t√©rmino general para los m√©todos de construcci√≥n de combinaciones de las variables para solucionar estos problemas sin dejar de describir los datos con suficiente precisi√≥n. **3. Investigaci√≥n/Selecci√≥n de Modelo - Clasificaci√≥n/Regresi√≥n**

1. Un modelo predictivo trata sobre el problema de aprender una funci√≥n de mapeo de entradas a salidas llamada aproximaci√≥n de funci√≥n.

1. **Plan:** esta es la parte del proyecto en la que est√° organizando las tareas, los horarios y configura las herramientas de gesti√≥n de proyectos.
2. **Code:** Aqu√≠ los desarrolladores de software est√°n desarrollando y revisando el c√≥digo.
3. **Build:** este paso es el primero hacia la automatizaci√≥n. El objetivo aqu√≠ es construir el c√≥digo fuente en un formato deseado, compilando, probando y desplegando en un lugar particular de la infraestructura.

5. **Release:** El c√≥digo ha pasado el proceso de prueba (integraci√≥n continua) y est√° listo para implementarse.
6. **Deploy:** El equipo operativo est√° implementando la nueva funci√≥n en producci√≥n. Pero como la automatizaci√≥n es uno de los principios de DevOps, es posible configurar una implementaci√≥n continua.
7. **Operate:** Las operaciones construyen o mantienen una infraestructura escalable
8. **Monitor:** El monitoreo es un paso importante ya que permite solucionar incidentes m√°s r√°pido y crear una mejor experiencia para su usuario final.
